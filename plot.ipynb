```python
import pandas as pd
import os
csv_dir = os.getcwd() + "/csv/mail.csv"
df = pd.read_csv(csv_dir, encoding = "latin", index_col=False)
```

```python
df.head(100)
```

```python
df["_D"] = pd.to_datetime(df["_D"], utc=True)
df.dtypes
```

```python
pd.DatetimeIndex(df["_D"])
```

```python
df["day"] = df["_D"].dt.day_name()
df["month"] = df["_D"].dt.month_name()
```

```python
import matplotlib.pyplot as plt 

plt.hist(df["day"], bins=30)
```

```python
plt.hist(df["month"])
```
